```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita')
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
```

```python
x = df[['year']]
y = df['per capita income (US$)']
```

```python
reg = linear_model.LinearRegression()
reg.fit(x,y)
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita')
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
plt.plot(df.year, reg.predict(df[['year']]), color='blue', marker='')
```
